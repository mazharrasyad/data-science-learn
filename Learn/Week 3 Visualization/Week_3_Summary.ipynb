```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.style.use('seaborn')

fig, ax = plt.subplots()
ax.plot(x, cos)
ax.plot(x, sin)
plt.show
```

```
plt.style.use('seaborn')

fig, ax = plt.subplots()
ax.plot(x, cos, linestyle='--', label='CosX')
ax.plot(x, sin, label='SinX')
ax.legend()
plt.show
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```
fig, ax = plt.subplots(figsize=(12,8))

rs_male = df[df['gender']=='male']['reading score']
ws_male = df[df['gender']=='male']['writing score']

rs_female = df[df['gender']=='female']['reading score']
ws_female = df[df['gender']=='female']['writing score']

# ax.scatter(df['reading score'], df['writing score'])
ax.scatter(rs_male, ws_male, color='b', label='Male')
ax.scatter(rs_female, ws_female, color='y', label='Female')

ax.set_xlabel('Reading Score')
ax.set_ylabel('Writing Score')
ax.set_title('Student Performance')
plt.legend()
plt.show()
```

```
df[df['gender']=='male']['reading score']
```

```
fig, ax = plt.subplots(figsize=(12,8))

MS = df['math score']
RS = df['reading score']
WS = df['writing score']

axmap = ax.scatter(RS, WS, c=MS, cmap='PuBuGn')

ax.set_xlabel('Reading Score')
ax.set_ylabel('Writing Score')
ax.set_title('Student Performance')
fig.colorbar(axmap)
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_csv('StudentsPerformance.csv')
```

```
df.head()
```

```
mean_score = df.groupby('parental level of education').mean()
```

```
mean_score
```

```
# x = mean_score.index
# x = np.arange(0, 12, 2)
x = np.arange(6)
y = mean_score['math score']
y2 = mean_score['reading score']
y3 = mean_score['writing score']
x
```

```
fig, ax = plt.subplots(figsize=(12, 8))
# ax.bar(x, y)
# ax.bar(x, y2)
# ax.bar([1, 2, 3, 4, 5, 10], y)
# ax.bar([1, 2, 3, 4, 5, 8], y2)
ax.bar(x, y, width=0.3)
ax.bar(x+0.3, y2, width=0.3)
ax.bar(x+0.6, y3, width=0.3)
plt.show()
```

```
x = mean_score.index
y = mean_score['math score']
y2 = mean_score['reading score']
y3 = mean_score['writing score']
```
