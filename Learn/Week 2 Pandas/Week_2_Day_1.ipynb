<h2>Quiz 3 : DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Perbedaan Series dengan list yaitu terletak pada indexnya, dimana jika list hanya bisa int sedangkan index pada series dapat diubah semisal menjadi string. Adapun dengan dictionary yaitu terletak pada duplikasi keynya yang dictionary tidak bisa duplikasi adanya key yang sama sedangkan series dapat melakukan hal tersebut.

Perbedaan dataframe dengan series terletak pada dimensinya, jika series 1 dimensi sedangkan dataframe 2 dimensi

```python
# code here
obj_series = pd.Series([99, 98, 97], index=['x', 'y', 'z'])
obj_series
```
